```python
import numpy as np
import pandas as pd
def data_preprocess(data):
    if 'OS(Y)' in data.columns:
        data['OS(d)'] = data['OS(Y)'] * 365
    if 'OS(M)' in data.columns:
        data['OS(d)'] = data['OS(M)'] * 30
    
    # map the stage/state into int if necessary
    stage_mapping = {'Stage IA': 1,'Stage IB': 1,'Stage IC': 1,'Stage IIA': 2,'Stage IIB': 2,'Stage IIC': 2,'Stage IIIA': 3,'Stage IIIB': 3,'Stage IIIC': 3,'Stage IV': 4}
    data['stage'] = data['stage'].map(stage_mapping)    
    state_mapping = {'Dead': 1,'Alive': 0}
    data['final_state'] = data['final_state'].map(state_mapping)
     
    return data
```
