```python
import pandas as pd
# GPL porbe file
probe_df = pd.read_csv("../data/GSE63885/GPL570-55999.txt", sep="\t", skiprows=16, header=0, dtype=str)  
# gene expression file
expression_df = pd.read_table("../data/GSE63885/GSE63885_series_matrix.txt", comment="!", header=0)  
```

```python
expression_df
```

```python
probe_df
```

```python
# merge these two datasets based on 'ID' and 'ID_REF'
merged_df = pd.merge(probe_df[['ID', 'Gene Symbol']], expression_df,
                     left_on='ID', right_on='ID_REF')
merged_df
```

```python
merged_df = merged_df.drop(columns=['ID','ID_REF'])
```

```python
merged_df
```

```python
merged_df = merged_df.rename(columns={'Gene Symbol': 'Gene'})
```

```python
merged_df[merged_df['Gene']=='PROM2']
```

```python
df_transposed = merged_df.set_index('Gene').transpose()

# set new index of the transposed data
df_transposed.reset_index(inplace=True)

# change the first colume name (the oringinal row name) into 'ID'
df_transposed.rename(columns={'index': 'ID'}, inplace=True)
df_transposed.columns.name = None
df_transposed
```

```python
df_transposed=df_transposed[['PROM1','PROM2']]
```

```python
df_transposed.to_csv('./OV/GSE140082/merged_gene_expression.csv', index=False)
```
